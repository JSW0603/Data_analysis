### Data cleaning

By looking at a random sample of the dataset rows, we observe that some entries in the columns like Installs have a few special characters due to the way the numbers have been represented. This prevents the columns from being purely numeric, making it difficult to use them in subsequent future mathematical calculations. Ideally, as their names suggest, we would want these columns to contain only digits from [0-9].

```python
char_to_remove = ["+", ",", "$", "M"]
cols_to_clean = ["Installs", "Price", "Size"]

for col in cols_to_clean:
    for char in char_to_remove:
        app[col] = app[col].apply(lambda x: x.replace(char, ""))

print("App filtered data info")
app.info()

print("App filtered data")
app.head(5)
```

```python
# In columns of Size, Installs and Price, There are values of string. So, we need to change them also

str_to_remove = ["Varies with device", "Free", "Everyone"]
cols_to_clean = ["Installs", "Price", "Size"]

for col in cols_to_clean:
    for string in str_to_remove:
        app[col] = app[col].apply(lambda x: x.replace(string, "0"))
```

### Changing data types

Through the info, we can check that the Dtypes of Size, Installs, Price is object. This is because these three columns originally had mixed input types: digits, string, and special characters.
We will work with columns of Installs, Size, Rating and Price. So, we need to work on Installs, Size, and Price to make them numeric.

```python
# Convert Three columns (Size, Installs, Price) to float data

app["Price"] = app["Price"].astype("float64")
app["Installs"] = app["Size"].astype("float64")
app["Price"] = app["Price"].astype("float64")

print(app.dtypes)
```
